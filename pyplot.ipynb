### Scatter Plot

The polyfit() function performs a least squares polynomial fit for the data it is given. 
A poly1d class is then created based on the array returned by polyfit(). 
This class is a one-dimensional polynomial class. It is a convenience class, used to 
encapsulate “natural” operations on polynomials. The poly1d object is then used to generate 
a set of values for use with the set of x values for the function pyplot.plot().

```python
import numpy as np
import matplotlib.pyplot as pyplot
x = (5, 5.5, 6, 6.5, 7, 8, 9, 10)
y = (120, 115, 100, 112, 80, 85, 69, 65)
# Generate the scatter plot
pyplot.scatter(x, y)
# Generate the trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
pyplot.plot(x, p(x), 'r')
# Display the figure
pyplot.show()
```

```python
import matplotlib.pyplot as plt

# Data
courses = ['Math', 'English', 'Physics', 'Chemistry', 'Art']
grades = [56, 78, 73, 69, 52]

# Create a pie chart
plt.figure(figsize=(6, 6))  # Adjust the figure size as needed
plt.pie(grades, labels=courses, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Grades by Course')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.show()
```

## Explanation

plt.figure(figsize=(6, 6)): This line creates a new figure for the pie chart. Here's what each part does:

plt.figure: This is a function from the matplotlib.pyplot module that creates a new figure or canvas for your plot.
figsize=(6, 6): This is an argument passed to the plt.figure function. It specifies the size of the figure in inches. In this case, (6, 6) sets the width and height of the figure to 6 inches each. You can adjust these values to control the size of the pie chart. Larger values will make the chart bigger, while smaller values will make it smaller.
plt.pie(grades, labels=courses, autopct='%1.1f%%', startangle=140): This line creates the pie chart itself, and it has several components:

grades: This is the data you want to visualize in the pie chart. It represents the grades in each course.
labels=courses: Here, you specify the labels for each slice of the pie chart. In this case, it uses the courses list, which contains the course names ('Math', 'English', 'Physics', 'Chemistry', 'Art') as labels for the corresponding grades.
autopct='%1.1f%%': This argument specifies the format of the autopct (automatic percentage) labels that appear on each slice. %1.1f%% formats the percentages with one decimal place. So, if a slice represents 25.4% of the total, it will be displayed as "25.4%".
startangle=140: The startangle parameter determines the angle at which the first slice starts. In this case, it starts at 140 degrees counterclockwise from the x-axis. This can be adjusted to change the initial orientation of the pie chart. A higher value would rotate the chart counterclockwise from the default starting position.
By combining these elements, you create a pie chart with the specified figure size, data, labels, percentage format, and starting angle. The plt.axis('equal') line ensures that the pie chart is displayed as a perfect circle, and plt.show() displays the chart on the screen.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with course and grade data
data = {'Course': ['Math', 'English', 'Physics', 'Chemistry', 'Art'],
        'Grade': [56, 78, 73, 69, 52]}
df = pd.DataFrame(data)

# Create a pie chart using the DataFrame
plt.figure(figsize=(6, 6))  # Adjust the figure size as needed
plt.pie(df['Grade'], labels=df['Course'], autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Grades by Course')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.show()
```

```python
# Bar Graph with Multiple or Grouped Categories
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with random data for multiple categories
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Category A': [30, 40, 25, 50, 35],
    'Category B': [20, 45, 10, 30, 50],
    'Category C': [10, 30, 15, 20, 40]
}

df = pd.DataFrame(data)

# Plot a stacked bar graph
plt.figure(figsize=(8, 6))
plt.bar(df['Year'], df['Category A'], label='Category A')
plt.bar(df['Year'], df['Category B'], bottom=df['Category A'], label='Category B')
plt.bar(df['Year'], df['Category C'], bottom=df['Category A'] + df['Category B'], label='Category C')

# Add labels and a legend
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Stacked Bar Graph with Multiple Categories')
plt.legend()

# Display the stacked bar graph
plt.tight_layout()
plt.show()
```

# # Sub plots

The matplotlib.pyplot.figure() function generates a Figure object, and you can work directly 
with this object to create multiple subplots within the same figure. This allows you to 
compare different views of the same data side by side. 
Here are two examples that demonstrate this concept:

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some example data
data1 = np.arange(0, 10)
data2 = np.arange(10, 20)

# Create a figure and two subplots side by side
fig = plt.figure(figsize=(10, 4))

# First subplot
ax1 = fig.add_subplot(121)  # 1 row, 2 columns, 1st subplot
ax1.plot(data1)
ax1.set_title('Data 1')

# Second subplot
ax2 = fig.add_subplot(122)  # 1 row, 2 columns, 2nd subplot
ax2.plot(data2)
ax2.set_title('Data 2')

# Set a title for the entire figure
fig.suptitle('Comparison of Two Datasets')

# Display the figure with both subplots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some example data
data1 = np.arange(0, 10)
data2 = np.arange(10, 20)
data3 = np.arange(20, 30)
data4 = np.arange(30, 40)

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
axs[0, 0].plot(data1)
axs[0, 0].set_title('Data 1')

# Second subplot
axs[0, 1].plot(data2)
axs[0, 1].set_title('Data 2')

# Third subplot
axs[1, 0].plot(data3)
axs[1, 0].set_title('Data 3')

# Fourth subplot
axs[1, 1].plot(data4)
axs[1, 1].set_title('Data 4')

# Set a title for the entire figure
fig.suptitle('Comparison of Four Datasets')

# Adjust layout
plt.tight_layout()

# Display the figure with all subplots
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create an industry-like DataFrame
data = {
    'Year': [2016, 2017, 2018, 2019, 2020],
    'Revenue': [800, 850, 900, 950, 1000],
    'Expenses': [600, 620, 650, 680, 700],
    'Profit': [200, 230, 250, 270, 300]
}
df = pd.DataFrame(data)

# Create a single figure with 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Subplot 1: Line plot for Revenue over Years
axs[0, 0].plot(df['Year'], df['Revenue'], marker='o', color='b')
axs[0, 0].set_title('Revenue Over Years')
axs[0, 0].set_xlabel('Year')
axs[0, 0].set_ylabel('Revenue')

# Subplot 2: Line plot for Expenses over Years
axs[0, 1].plot(df['Year'], df['Expenses'], marker='o', color='r')
axs[0, 1].set_title('Expenses Over Years')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('Expenses')

# Subplot 3: Bar plot for Profit over Years
axs[1, 0].bar(df['Year'], df['Profit'], color='g')
axs[1, 0].set_title('Profit Over Years')
axs[1, 0].set_xlabel('Year')
axs[1, 0].set_ylabel('Profit')

# Subplot 4: Stacked bar plot for Revenue and Expenses over Years
axs[1, 1].bar(df['Year'], df['Revenue'], color='b', label='Revenue')
axs[1, 1].bar(df['Year'], df['Expenses'], color='r', label='Expenses', bottom=df['Revenue'])
axs[1, 1].set_title('Revenue and Expenses Over Years')
axs[1, 1].set_xlabel('Year')
axs[1, 1].set_ylabel('Amount')
axs[1, 1].legend()

# Set a title for the entire figure
fig.suptitle('Financial Analysis Over the Years')

# Adjust layout
plt.tight_layout()

# Display the figure with all subplots
plt.show()
```

Now, let's break down the code step by step:

Select the Subplot: axs[0, 0] selects the first subplot in the 2x2 grid. This subplot will be used to create the line plot for Revenue over Years.

Create the Line Plot: axs[0, 0].plot(df['Year'], df['Revenue'], marker='o', color='b') creates the line plot. Here's what each part does:

df['Year'] is the x-axis data, representing the years.
df['Revenue'] is the y-axis data, representing the revenue values.
marker='o' specifies that data points on the line should be marked with circles ('o').
color='b' sets the color of the line to blue ('b').
Set Title: axs[0, 0].set_title('Revenue Over Years') sets the title for this subplot to "Revenue Over Years."

Set X-Axis Label: axs[0, 0].set_xlabel('Year') adds a label to the x-axis, specifying that it represents the "Year."

Set Y-Axis Label: axs[0, 0].set_ylabel('Revenue') adds a label to the y-axis, indicating that it represents "Revenue."

The result of this code is a line plot in the first subplot, showing the revenue data over the years. Data points are marked with circles, and the plot is labeled with a title and axis labels for clarity.

You can follow a similar approach to create the other subplots, adjusting the data and properties as needed to represent different aspects of your data.


axs is a 2D array or grid of subplots created by plt.subplots. In this context, axs[0, 1] is a reference to a specific subplot within this grid. The two numbers inside the square brackets represent row and column indices, respectively.

The first number (0) refers to the row index, where 0 is the index of the first row in the grid.

The second number (1) refers to the column index, where 1 is the index of the second column in the grid.

So, axs[0, 1] specifically refers to the subplot located in the first row (row index 0) and the second column (column index 1) of the subplot grid. This allows you to access and modify the properties of that particular subplot.
