```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# Sample data with non-zero proportions for each category combination
data = pd.DataFrame({
    'Weather': ['Sunny', 'Rainy', 'Sunny', 'Sunny', 'Rainy', 'Rainy', 'Sunny', 'Rainy', 'Sunny', 'Sunny'],
    'Location': ['Outdoor', 'Indoor', 'Indoor', 'Outdoor', 'Outdoor', 'Outdoor', 'Indoor', 'Indoor', 'Outdoor', 'Outdoor'],
    'Activity': ['Running', 'Reading', 'Running', 'Running', 'Reading', 'Reading', 'Running', 'Reading', 'Running', 'Running'],
})

# Create the mosaic plot
mosaic(data, ['Weather', 'Location', 'Activity'], title='Mosaic Plot')

# Create a table to display the data
fig, ax = plt.subplots()
ax.axis('off')
ax.table(cellText=data.values, colLabels=data.columns, cellLoc='center', loc='center', colColours=['palegreen'] * len(data.columns))
plt.title('Data Table')

# Show the plots
plt.show()
```
