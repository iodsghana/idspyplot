```python
import matplotlib.pyplot as plt

# Replicating the data
data = {
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Age Group': ['18-30', '18-30', '31-50', '31-50', '51+', '51+'],
    'Preference': ['Like', 'Dislike', 'Like', 'Like', 'Dislike', 'Like']
}

# Creating a DataFrame (assuming you're using Pandas)
import pandas as pd
df = pd.DataFrame(data)
print(df)
# Creating a contingency table using crosstab
contingency_table = pd.crosstab(index=df['Age Group'], columns=[df['Gender'], df['Preference']])

# Visualizing with stacked bar chart
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Customer Preferences by Gender and Age Group')
plt.ylabel('Count')
plt.xlabel('Age Group')
plt.xticks(rotation=0)
plt.legend(title='Gender & Preference')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Generating sample data
categories = ['Electronics', 'Clothing', 'Books', 'Home Appliances']
data = []

for i in range(100):
    category = random.choice(categories)
    data.append({'Category': category})

# Creating a DataFrame
df = pd.DataFrame(data)

# Counting the frequency of each category
category_counts = df['Category'].value_counts()
print(category_counts)
```

```python
print(category_counts)
```

```python
# Creating the bar chart for frequency
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Frequency of Customer Preferences for Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Creating the bar chart for proportion (normalized)
proportions = category_counts / category_counts.sum()

plt.figure(figsize=(10, 6))
proportions.plot(kind='bar')
plt.title('Proportion of Customer Preferences for Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Generating sample data
categories = ['Electronics', 'Clothing', 'Books', 'Home Appliances']
data = []

for _ in range(100):
    category = random.choice(categories)
    data.append({'Category': category, 'Gender': random.choice(['Male', 'Female'])})

# Creating a DataFrame
df = pd.DataFrame(data)

# Counting the frequency of each category and gender combination
gender_counts = df.groupby(['Category', 'Gender']).size().unstack()

# Creating the stacked bar chart
plt.figure(figsize=(10, 6))
gender_counts.plot(kind='bar', stacked=True)
plt.title('Customer Preferences for Product Categories by Gender')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
